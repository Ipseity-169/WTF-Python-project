**Project Overview:**
We wanted to build a project that encompasses all that we have been learning in class. We decided on a system that allows users to buy ticket for available upcoming concerts. On our system, users will be able to check for upcoming concerts, see if they are available, reserve a seat, get a discount if they enter the promo code and get a reciept for their purchase. Users can also decide the type of ticket they want, from the options available: Regular, VIP, VVIP and Table of Ten.

Using OOP, all our coding was written under a single class, and we created different functions to perform specific tasks. We made use of our knowlege of loops, conditional statements, and dictionaries in order to make our code more concise and legible. Also, to catch and  handle exceptions in our code, we used 'try' and 'except' statements.











```python
import random
class Concert_Ticket:
    def __init__(self, name, dict1, dict2):
        self.name = name
        self.dict1=dict1
        self.dict2=dict2
        
        
###############################################################        
    def show_menu(self): #Function to return list of operations available
        print(f"Welcome {self.name},")
        print("1- Check available concert \t 2- Reserve seat")
        print("3- Pay  \t 4- Exit") 
        try: 
            option = int(input("Enter Your Option: "))
        except:
            print("Invalid Option")
            self.end()
        else:
            if option == 1:
                self.available(dict1)
            elif option == 2:
                self.reserve(dict1)
            elif option == 3:
                self.pay(dict2)
            elif option == 4:
                self.exit()
            else:
                print("Invalid Option")
                self.end()
        

################################################################
    def available(self, dict1):

     # This functions return a table of upcoming concerts
        print('')
        print(f"{'Available concerts':<60}{'Concert date':<20}{'Ticket remaining':>10}")
        for key in self.dict1:
            print(f'{key: <40} {self.dict1[key][0]:>35}, {self.dict1[key][1]:>10} tickets')
        self.end()
        
        
###############################################################        
    def reserve(self, dict1): # Function to check whether a concert is still available or is sold out
        for index, key in enumerate (self.dict1):
            print(f'Please select {index} for {key}')
            
        try:
            option=int(input("Please enter your choice here:"))
            keys=list(self.dict1)[option]
        except:
            print("invalid option")
            self.available(dict1)
        else:
            if dict1[keys][1]==0:
                print(f"{keys} is sold out")
                self.end()
            else:
                print(f'You have selected {keys}')
            self.end()
            
###############################################################   
    def pay(self, dict2): # Function to enable payment in the system
        print('')
        print(f"{'Available Tickets':<25}{'Ticket class':>10}{'Price in Naira':>20}")
        for index, key in enumerate (self.dict2):
            print(f'Please select {index}{key:>20}{dict2[key]:>20}')
            
        try:
            option=int(input("Please enter your choice here:"))
            keys=list(self.dict2)[option]
        except:
            print("invalid option")
            self.end()
        else:
            keys=list(self.dict2)[option]
            print(f'You have selected {keys}, the price is {dict2[keys]}')
            
            Total_price = self.dict2[keys]
            price= self.promo_code(Total_price)
            print("")
            amount = float(input("Enter your amount for deposite: "))
            if amount < price:
                print("Insufficient payment!")
                self.end()
            elif amount > price:
                Change=amount - price
                print(f'Your change is', Change)
                self.receipt(dict2, price, keys)
            else:
                self.receipt(dict2, price, keys)
#####################################################################

    def promo_code(self, Total_price): # Function to return a discounted price when promo code is entered

        print('')
        try:
            promo = int(input("Promo code:"))
        except: 
            print(f'invalid code, sorry no discount \nYour ticket price is', Total_price)
            return Total_price
        else:
            if promo == 123:
                N_price=Total_price - (Total_price *0.3)
                print(f"Congratulation you have earned 30% discount! \nYour ticket price is now", N_price)
                return N_price 
            else:
                print(f'invalid code, sorry no discount \nYour ticket price is', Total_price)
                return Total_price
######################################################################
    def receipt(self, dict2, price, key):
      # Prints out a receipt after payment
        Ticket_id=str(random.randint(10000, 99999))
        Ticket_type=key
        Amount_paid=price
        print("Transaction successful, Enjoy your show")
        print("")
        print(f"Thank you {self.name} for transacting with us!\nYour ticket number is {Ticket_id}")
        print(f"Your ticket type is {Ticket_type}")
        print(f"Total amount paid is {Amount_paid}")
        self.end()
        




#####################################################################            
    def end(self): # Function to signals the end of any operation
        print("")
        print("Do you want to perform another Operation?")
        answer=input("(y) for yes and (n) or any other key for no:")
        if answer=='y' or answer=='yes' or answer=='Yes' or answer=='YES' or answer=='YEs' or answer=='YeS':
            self.show_menu()
        else:
            self.exit()
            
            
####################################################################           
    def exit(self):
        print("Thank you for transacting with us")
```
